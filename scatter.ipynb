```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('tips')
df.head()
```

### Matplotlip

```python
plt.style.use('ggplot')
plt.figure(figsize=(10,6))
plt.title('Reletionship between total bill vs tip')
plt.scatter(df['total_bill'], df['tip'])
plt.xlabel('total bill')
plt.ylabel('tip')
plt.show()
```

**Try Replicate Hue parameter in seaborn using matplotlib**

```python
df_male = df[df['sex']=='Male']
df_female = df[df['sex']=='Female']
```

```python
plt.style.use('seaborn')
plt.figure(figsize=(10,6))
plt.title('Reletionship between total bill vs tip')
plt.scatter(df_male['total_bill'],
            df_male['tip'],
            label='male customer')
plt.scatter(df_female['total_bill'],
            df_female['tip'],
            label='female customer')
plt.xlabel('total_bill')
plt.ylabel('tip')
plt.legend()
plt.show()
```

### Seaborn

```python
plt.style.use('seaborn')
plt.title('Reletionship between total bill vs tip')
sns.scatterplot(data=df,y='tip',x='total_bill')
plt.show()
```

**Scatter group**

```python
plt.title('Reletionship between total bill vs tip')
sns.scatterplot(data=df,y='tip',x='total_bill',hue='sex')
```

```python
plt.title('Relationship between total bill vs tip')
sns.scatterplot(data=df,
                y='tip',
                x='total_bill',
                hue='sex',
                style='time')
```
