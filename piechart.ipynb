```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = sns.load_dataset('tips')
df.head()
```

### Matplotlib

```python
df_sex = df.reset_index()
```

```python
df_sex = df.reset_index()
gender_groupby = df_sex[['sex','index']].groupby('sex').count().reset_index()
gender_groupby
```

```python
# rename nama index
gender_count = df['sex'].value_counts().reset_index().rename(columns={
    'index':'sex',
    'sex':'count'
})
gender_count
```

```python
df.head()
```

```python
plt.figure(figsize=(8,6))
plt.pie(gender_count['count'],
        labels=gender_count['sex'])
plt.axis('equal')
plt.title('Comparison between male and female customers')

plt.show()
```

**startangle**

```python
plt.figure(figsize=(8,6))
plt.pie(gender_count['count'],
        labels=gender_count['sex'],
        startangle=90)
plt.axis('equal')
plt.title('Comparison between male and female customers')
plt.show()
```

**counterclock**

```python
plt.figure(figsize=(8,6))
plt.pie(gender_count['count'],
        labels=gender_count['sex'],
        startangle=90,
        counterclock=False)
plt.axis('equal')
plt.title('Comparison between male and female customers')
plt.show()
```

**explode**

```python
plt.figure(figsize=(8,6))
plt.pie(gender_count['count'],
        labels=gender_count['sex'],
        startangle=90,
        counterclock=False,
        explode=[0.1,0])
plt.axis('equal')
plt.title('Comparison between male and female customers')
plt.show()
```

### **Piechart styling**

```python
plt.figure(figsize=(8,6))
plt.pie(gender_count['count'],
        labels=gender_count['sex'],
        autopct='%.1f%%',
        colors=['#3E54AC','#FFACAC'],
        startangle=90,
        counterclock=False)
plt.axis('equal')
plt.title('Comparison between male and female customers')

plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.pie(gender_count['count'],
        labels=gender_count['sex'],
        autopct='%.1f',
        colors=['#3E54AC','#FFACAC'],
        startangle=90,
        counterclock=False,
        textprops={'color':'brown'})
plt.axis('equal')
plt.title('Comparison between male and female customers')

plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.pie(gender_count['count'],
        labels=gender_count['sex'],
        autopct='%.1f',
        colors=['#3E54AC','#FFACAC'],
        startangle=90,
        counterclock=False)
fig = plt.gcf()
centre_circle = plt.Circle((0,0),0.45,fc='white')
fig.gca().add_artist(centre_circle)
plt.axis('equal')
plt.title('Comparison between male and female customers')
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.pie(gender_count['count'],
        labels=gender_count['sex'],
        autopct='%.1f%%',
        colors=['#3E54AC','#FFACAC'],
        startangle=90,
        counterclock=False)
fig= plt.gcf()
centre_circle = plt.Circle((0,0), 0.75, fc='white')
fig.gca().add_artist(centre_circle)
plt.axis('equal')
plt.title('Comparison between male and female customers')
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.pie(gender_count['count'],
        labels=gender_count['sex'],
        autopct='%.1f%%',
        colors=['#3E54AC','#FFACAC'],
        startangle=90,
        counterclock=False,
        explode=[0.02,0])
fig = plt.gcf()
centre_circle = plt.Circle((0,0), 0.75, fc='white')
fig.gca().add_artist(centre_circle)
plt.axis('equal')
plt.title('Comparison between male and female customers')
plt.show()
```
