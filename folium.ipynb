## Marker

```python
# buat peta jakarta
jakarta_map = folium.Map(location=[-6.200000, 106.816666], zoom_start=12)

# buat masker Monas
monas_pin = folium.Marker(
    location=[-6.176132,106.822864],
    tooltip='Monumen Nasional',
    popup='Monas',
    icon= folium.Icon(
        color='red',
        icon='home'
    )
)
folium.Marker([-6.1376, 106.8171],
              popup='Kota Tua Jakarta',
              tooltip='Mari ke Kota Tua Jakarta',
              icon=folium.Icon(color='blue', icon='cloud')).add_to(jakarta_map)

ancol_area = folium.CircleMarker(
    location=[-6.128847, 106.833053],
    tooltip='Ancol',
    popup='Area rawan banjir',
    radius=30,
    color='blue'
)

monas_pin.add_to(jakarta_map)
monas_pin.add_to(jakarta_map)

jakarta_map
```
