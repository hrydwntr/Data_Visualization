```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = sns.load_dataset('penguins')
df.head()
```

```python
correlation = df.corr()
correlation
```

```python
sns.heatmap(correlation)
# - Value 0 - 0,3, berarti lemah
# - Value 0,3 - 0,7 berarti hubungannya sedang
# - Value 0,7 - 1 berarti kekuatan hubungan kuat 
```

**styling Heatmap**

```python
sns.heatmap(correlation, annot=True)
```

```python
sns.heatmap(correlation, annot=True, fmt='.1f')
```

```python
sns.heatmap(correlation, linewidths=.75,annot=True, fmt='.1f')
```

```python
sns.heatmap(correlation,linewidths=.75,
            annot=True, fmt='.1f',
            cmap='rocket_r')
```

```python
sns.heatmap(correlation, linewidths=.75,
            annot=True, fmt='.1f',
            cmap='crest')
```

```python
sns.heatmap(correlation, linewidths=.75,
            annot=True, fmt='.1f',
            cmap='crest',
            vmin=-1,vmax=1)
```

```python
sns.heatmap(correlation, linewidths=.75,
            annot=True, fmt='.1f',
            cmap='coolwarm',
            vmin=-1,vmax=1)
```

```python
plt.figure(figsize=(5,5))

mask = np.triu(np.ones_like(correlation))

sns.heatmap(correlation, cmap='coolwarm',
            mask=mask, linewidths=.5, annot_kws={'size':10},
            annot=True, fmt='.1f',
            vmin=-1, vmax=1)
plt.show()
```
