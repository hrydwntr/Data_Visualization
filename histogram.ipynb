```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = sns.load_dataset('tips')
df.head()
```

## Matplotlib 

```python
plt.figure(figsize=(8,5))
plt.hist(df['tip'])
plt.title('Distribution of Tip')
plt.ylabel('Frequency') #just name
plt.xlabel('Tip') #just name
# plt.savefig('')
plt.show()
```

```python
plt.figure(figsize=(5,2), dpi=250)
plt.grid(True, alpha=0.3)
plt.hist(df['tip'], 20, color='lightgreen', edgecolor='black')
plt.title('Distribution of Tip')
plt.xlabel('Tip')
plt.ylabel('Frequency')
plt.legend(['Tip distribution'])
plt.show()
```

## Seaborn

```python
plt.figure(figsize=(10,5))
plt.title('Histogram Tip')
sns.histplot(df['tip'])
plt.ylabel('freq')
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.title('Histogram Tip')
sns.histplot(data=df,x='tip',kde=True, hue='sex')
plt.ylabel('freq')
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.title('Histogram Tip')
sns.histplot(data = df, x='tip', kde=True, hue='sex', multiple='stack')
plt.ylabel('freq')
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.title('Histogram Tip')
sns.histplot(data=df,
             x='tip',
             kde=True,
             hue='sex',
             multiple='stack',
             element='step')
plt.ylabel('freq')
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.title('Histogram Tip')
sns.histplot(data=df,
             x='tip',
             kde=True,
             hue='sex',
             multiple='stack',
             element='poly')
plt.ylabel('freq')
plt.show()
```
