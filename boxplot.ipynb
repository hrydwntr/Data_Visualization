```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = sns.load_dataset('tips')
df.head()
```

## Matplotlip

```python
plt.figure(figsize=(8,5))
plt.boxplot(df['tip'])
plt.title('Boxplot of Tip')
plt.xlabel('Tip')
plt.ylabel('Value')
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.boxplot(df['tip'], vert=False)
plt.title('Boxplot of Tip')
plt.xlabel('value')
plt.ylabel('Tip')
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.boxplot(df['tip'])
plt.grid(True)
plt.title('Boxplot of Tip')
plt.xlabel('Value')
plt.ylabel('Tip')
plt.show()
```

```python
plt.style.use('grayscale')
plt.figure(figsize=(8,5))
plt.boxplot(df['tip'])
plt.grid(True)
plt.title('Boxplot of Tip')
plt.xlabel('Value')
plt.ylabel('tip')
plt.show()
```

```python
plt.style.use('seaborn')
plt.figure(figsize=(8,5))
plt.boxplot(df['tip'])
plt.grid(True)
plt.title('Boxplot of Tip')
plt.xlabel('value')
plt.ylabel('tip')
plt.show()
```

```python
plt.style.use('ggplot')
plt.figure(figsize=(8,5))
plt.boxplot(df['tip'])
plt.grid(True)
plt.title('Boxplot of Tip')
plt.xlabel('value')
plt.ylabel('TIp')
plt.show()
```

```python
plt.style.use('dark_background')
plt.figure(figsize=(8,5))
plt.boxplot(df['tip'])
plt.grid(True)
plt.title('Boxplot of Tip')
plt.xlabel('value')
plt.ylabel('TIp')
plt.show()
```

## Seaborn

```python
plt.style.use('default')
plt.title('Boxplot of Tip')
sns.boxplot(data=df,x='tip')
plt.show()
```

```python
plt.style.use('default')
plt.title('Boxplot of Tip')
sns.boxplot(data=df, y='tip')
plt.show()
```

```python
plt.style.use('default')
plt.title('Boxplot of Tip')
sns.boxplot(data=df, x="sex", y="tip")
plt.show()
```
