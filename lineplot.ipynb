```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
sns.__version__
```

```python
df = sns.load_dataset('flights')
df.head()
```

### Matplotlip

```python
year = df[df['month']=='Jan']['year']
passanger = df[df['month']=='Jan']['passengers']
passanger
```

```python
plt.plot(year,passanger)
```

```python
plt.figure(figsize=(8,4))
plt.plot(year,passanger,'-o')
plt.title('Line of passenger flight')
plt.xlabel('year')
plt.ylabel('passanger')
```

```python
plt.figure(figsize=(8,5))
plt.plot(year,passanger, 'r:o')
plt.title('Line of passenger flight')
plt.xlabel('year')
plt.ylabel('passanger')
```

```python
plt.figure(figsize=(8,4))
plt.plot(year, passanger, 'ro')
plt.title('Line of passenger flight')
plt.xlabel('year')
plt.ylabel('passenger')
```
